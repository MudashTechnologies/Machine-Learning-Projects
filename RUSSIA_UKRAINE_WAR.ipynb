```python
X_train_series = X_train.values.reshape((X_train.shape[0], X_train.shape[1], 1))
X_valid_series = X_valid.values.reshape((X_valid.shape[0], X_valid.shape[1], 1))
print('Train set shape', X_train_series.shape)
print('Validation set shape', X_valid_series.shape)
```

### CNN

```python
model_cnn = Sequential()
model_cnn.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(X_train_series.shape[1], X_train_series.shape[2])))
model_cnn.add(MaxPooling1D(pool_size=2))
model_cnn.add(Flatten())
model_cnn.add(Dense(50, activation='relu'))
model_cnn.add(Dense(1))
model_cnn.compile(loss='mse', optimizer=adam)
model_cnn.summary()
```

```python
cnn_history = model_cnn.fit(X_train_series, Y_train, validation_data=(X_valid_series, Y_valid), epochs = 40, verbose=1)
```

```python
import matplotlib.pyplot as plt 
loss_train = cnn_history.history['loss']
loss_val = cnn_history.history['val_loss']
epochs = range(1,41)
plt.plot(epochs, loss_train, 'r', label='Training loss')
plt.plot(epochs, loss_val, 'g', label='validation loss')
plt.title('Training and Validation loss for Adam')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

### LSTM

```python
model_lstm = Sequential()
model_lstm.add(LSTM(50, activation='relu', input_shape=(X_train_series.shape[1], X_train_series.shape[2])))
model_lstm.add(Dense(1))
model_lstm.compile(loss='mse', optimizer=adam)
model_lstm.summary()
```

```python
lstm_history = model_lstm.fit(X_train_series, Y_train, validation_data=(X_valid_series, Y_valid), epochs = 40, verbose=1)
```

```python
import matplotlib.pyplot as plt 
loss_train = lstm_history.history['loss']
loss_val = lstm_history.history['val_loss']
epochs = range(1,41)
plt.plot(epochs, loss_train, 'r', label='Training loss')
plt.plot(epochs, loss_val, 'g', label='validation loss')
plt.title('Training and Validation loss for Adam')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

### CNN-LSTM 

```python
from tensorflow.keras.layers import Bidirectional, Dropout
```

```python
cnn_lstm_model = Sequential()
cnn_lstm_model.add(Conv1D(filters=256, kernel_size=2, activation='relu', input_shape=(X_train_series.shape[1], X_train_series.shape[2])))
cnn_lstm_model.add(MaxPooling1D(pool_size=2))
cnn_lstm_model.add(Flatten())
cnn_lstm_model.add(RepeatVector(30))
cnn_lstm_model.add(LSTM(units=50, return_sequences=True, activation='relu'))
cnn_lstm_model.add(Bidirectional(LSTM(128, activation='relu')))
cnn_lstm_model.add(Dense(50, activation='relu'))
cnn_lstm_model.add(Dense(1))
cnn_lstm_model.compile(loss='mse', optimizer='adam')
```

```python
cnn_lstm_history = cnn_lstm_model.fit(X_train_series, Y_train, validation_data=(X_valid_series, Y_valid), epochs = 40, verbose=1)
```

```python
import matplotlib.pyplot as plt 
loss_train = cnn_lstm_history.history['loss']
loss_val = cnn_lstm_history.history['val_loss']
epochs = range(1,41)
plt.plot(epochs, loss_train, 'r', label='Training loss')
plt.plot(epochs, loss_val, 'g', label='validation loss')
plt.title('Training and Validation loss for Adam')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
